```python
file1 = 'pheno-master-data-collection-v1-prod-eu-west-1_v11_before_2023_False_show_feature_True.csv' # eu
file2 = 'pheno-master-data-collection-v1-prod-us-east-1_v11_before_2023_False_show_feature_True.csv' # us

df3 = pd.read_csv(f's3://ds-users/mariag/data_counts/{file1}')
df3.set_index('dataset', inplace=True)

df4 = pd.read_csv(f's3://ds-users/mariag/data_counts/{file2}')
df4.set_index('dataset', inplace=True)

df_merge = df3[['first_date', 'last_date', 'count_col_notna', 'nunique_participants_not_na']].merge(
    df4[['first_date', 'last_date', 'count_col_notna', 'nunique_participants_not_na']],
    left_index=True, 
    right_index=True, 
    suffixes=('_eu', '_us'))
df_merge['count_delta'] = df_merge['count_col_notna_us'] - df_merge['count_col_notna_eu']
df_merge['participants_delta'] = df_merge['nunique_participants_not_na_us'] - df_merge['nunique_participants_not_na_eu']
df_merge.rename(columns={'count_col_notna_eu': 'count_eu', 'nunique_participants_not_na_eu': 'participants_eu', 
                         'count_col_notna_us': 'count_us', 'nunique_participants_not_na_us': 'participants_us'}, inplace=True)
cols_to_show = ['last_date_eu', 'last_date_us', 'count_eu', 'participants_eu', 
                'count_us', 'participants_us','count_delta', 'participants_delta']
df_merge[cols_to_show].to_csv('s3://ds-users/mariag/data_counts/data_counts_diff.csv')
df_merge[cols_to_show]
```

```python
list_of_count_file_names = [
    'pheno-academic-collection-v1-prod-eu-west-1_v11_before_2023_False_show_feature_True.csv',
    'pheno-academic-collection-v1-prod-us-east-1_v11_before_2023_False_show_feature_True.csv',
    'pheno-master-data-collection-v1-prod-eu-west-1_v11_before_2023_False_show_feature_True.csv',
    'pheno-master-data-collection-v1-prod-us-east-1_v11_before_2023_False_show_feature_True.csv',
    'pheno-master-data-collection-v1.2-prod-eu-west-1_v12_before_2023_False_show_feature_True.csv',
    'pheno-academic-collection-v1.2-prod-eu-west-1_v12_before_2023_False_show_feature_True.csv',
]

base_path = 's3://ds-users/mariag/data_counts/'
```

```python
df_col_order = df = pd.read_csv(base_path + list_of_count_file_names[-2])
index_order = df_col_order.dataset.to_list()
```

```python
# cols_to_show = ['last_date', 'count_col_notna', 'nunique_participants_not_na']
cols_to_show = ['nunique_participants_not_na']

df_all = None
for file in list_of_count_file_names:
    print(f'Reading {file}')
    df = pd.read_csv(base_path + file)
    df.set_index('dataset', inplace=True)
    df = df[cols_to_show]
    file = file.replace('_before_2023_False_show_feature_True.csv', '')
    df.rename(columns={'last_date': f'last_date_{file}', 
                       'count_col_notna': f'count_col_notna_{file}',
                       'nunique_participants_not_na': f'{file}'}, inplace=True)
    if df_all is None:
        df_all = df
    else: 
        df_all = df_all.merge(df, left_index=True, right_index=True, how='outer')
```
